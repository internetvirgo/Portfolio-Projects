```python
import pandas as pd
import os
```

```python
data_import = pd.read_csv('data_cleaning_challenge.csv')
data_import
```

```python
drop_extra = data_import.drop(columns=["Unnamed: 9", "Unnamed: 10"])
drop_extra
```

```python
drop_nan = drop_extra[drop_extra["Row Type"].notna()]
drop_nan
```

```python
column_values =[]
counter = 0

for i in drop_nan["Row Type"]:
    if "first name" in i:
        counter += 1
    column_values.append(counter)

column_values
```

```python
iter_cols = drop_nan
iter_cols["Iteration"] = column_values 
iter_cols
```

```python
drop_extra_column_name_row = iter_cols[iter_cols["Row Type"] != "Row Type"]
drop_extra_column_name_row
```

```python
name_dataframe = drop_extra_column_name_row[drop_extra_column_name_row["Row Type"].str.contains("first name")]
name_dataframe.drop(columns=["Speed1", "Speed2", "Electricity", "Effort", "Weight", "Torque"], inplace=True)
name_dataframe.rename(columns={"Row Type": "First Name", "Iter Number": "Last Name", "Power1": "Date"}, inplace=True)
name_dataframe
```

```python
name_dataframe["First Name"] = name_dataframe["First Name"].str[12:]
name_dataframe["Last Name"] = name_dataframe["Last Name"].str[11:]
name_dataframe["Date"] = name_dataframe["Date"].str[6:]
name_dataframe
```

```python
no_name_dataframe = drop_extra_column_name_row[~drop_extra_column_name_row["Row Type"].str.contains("first name")]
no_name_dataframe
```

```python
cleaned_data_set = pd.merge(left = name_dataframe, right = no_name_dataframe, how="inner", on="Iteration")
cleaned_data_set
```

```python
cleaned_data_set.drop(columns=["Iteration"], inplace=True)
cleaned_data_set
```

```python
cleaned_data_set.to_csv(r'Downloads\cleaned_data_set.csv', index = False)
```
